```python
import pandas as pd

# Sample product data
data = {
    'Product': ['Laptop', 'Smartphone', 'Headphones', 'Monitor', 'Keyboard', 'Mouse', 'Printer', 'Webcam', 'Speaker', 'Tablet'],
    'Price': [1200, 800, 150, 300, 50, 30, 200, 70, 100, 400],
    'InStock': [15, 30, 50, 20, 100, 80, 10, 25, 40, 18]
}

sample_df = pd.DataFrame(data)
sample_df.head()
```

```python
# Import matplotlib for visualization
import matplotlib.pyplot as plt

# Scatter plot: Product vs InStock
plt.figure(figsize=(10,6))
plt.scatter(sample_df['Product'], sample_df['InStock'], color='teal', s=100)
plt.title('Stock Quantity by Product')
plt.xlabel('Product')
plt.ylabel('In Stock')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
